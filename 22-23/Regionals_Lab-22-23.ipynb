# Task 5: Cross-validating model perfomance

## Subtask 1 : Define a function for cross-validation

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

def cross_validate(X, Y):
    # Create a Logistic Regression object
    model = LogisticRegression()

    # Perform k-fold cross-validation
    scores = cross_val_score(model, X, Y, cv=5, scoring='recall')

    # Return the average recall rate
    return scores.mean()
````

## Subtask 2: Compare model perfomance

````python

# Task 2: Compare model performance
# Original dataset
average_recall_original = cross_validate(X, Y)
print(f'\n\nAverage recall rate (original dataset): {average_recall_original:.2f}\n\n')

# Undersampled dataset
average_recall_undersample = cross_validate(X_undersample, Y_undersample)
print(f'\n\nAverage recall rate (undersampled dataset): {average_recall_undersample:.2f}\n\n')

# Oversampled dataset
average_recall_oversample = cross_validate(X_oversample, Y_oversample)
print(f'\n\nAverage recall rate (oversampled dataset): {average_recall_oversample:.2f}\n\n')
````
