```python
high_income_education_distribution = df[df['salary'] == '>50K']['education'].value_counts()

# List of education levels considered at least high school
high_school_or_above = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']

# Check if there is any education level below high school in the high income group
at_least_high_school_education = high_income_education_distribution.loc[high_school_or_above].sum() == high_income_education_distribution.sum()

if at_least_high_school_education:
    print("It is true that people who earn more than 50K have at least high school education.")
else:
    print("Not all people who earn more than 50K have at least high school education.")
```

###  9) Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

```python
import pandas as pd
# load the dataset from the URL
file = 'adult_data.csv'

columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary']
df = pd.read_csv(file, names=columns)
# define the list of marital statuses that are considered married
married = ["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"]
# create a new column that indicates whether a person is married or single
df["marital-status"] = df["marital-status"].apply(lambda x: "Married" if x in married else "Single")
# create a new column that indicates whether a person earns more than 50K
df["salary"] = df["salary"].apply(lambda x: 1 if x == ">50K" else 0)
# group by race and marital status and then calculate the mean of salary
df.groupby(by=["race", "marital-status"])["salary"].mean()
```

### 10) What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python
import pandas as pd

file = 'adult_data.csv'

columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary']
df = pd.read_csv(file, names=columns)

# find the maximum number of hours a person works per week
max_hours = df["hours-per-week"].max()

# count the number of people who work the maximum number of hours
count = df["hours-per-week"].value_counts()[max_hours]

# create a new column that indicates whether a person earns more than 50K
df["salary"] = df["salary"].apply(lambda x: 1 if x == ">50K" else 0)

# calculate the percentage of people who earn more than 50K and work the maximum number of hours
percentage = df[df["hours-per-week"] == max_hours]["salary"].mean() * 100

# print the results
print(f"The maximum number of hours a person works per week is {max_hours}")
print(f"There are {count} people who work such a number of hours")
print(f"The percentage of those who earn more than 50K among them is {percentage:.1f}%")
```
